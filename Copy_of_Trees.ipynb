Binary Tree


is a special datastructure used for data storage purposes. each node can have a maximum of two children.

 Basic Operations
1.   insertion
2.   search
3.   traversal (Pre-order, inorder, postorder)

## Terminologies
root, node, edge ,height of a node, depth of a node, degree of a node, forest

```
#bi_trees
class Node:
  def __init__(self, data): #root
    self.left= None
    self.data= data
    self.right= None

  def insert(self, data):#how to insert a node
    if self.data:

      if data < self.data:#left child
        if self.left is None:
          self.left = Node(data)

        else:
          self.left.insert(data)

      else:#right child
        if self.right is None:
          self.right = Node(data)

        else:
          self.right.insert(data)

    else:
      self.data = data


#print the tree
  def PrintTree(self):
    if self.left:
      self.left.PrintTree()

    print(self.data)

    if self.right:
      self.right.PrintTree()

#call the funtions
root = Node(10)

root.insert(6)
root.insert(8)
root.insert(4)
root.insert(15)
root.insert(21)
root.insert(18)


root.PrintTree()
```
